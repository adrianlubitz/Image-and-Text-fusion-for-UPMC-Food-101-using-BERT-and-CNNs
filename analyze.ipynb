# Calculate Accuracy for Image Model

## Load Model

```python
image_model = keras.models.load_model('inception_model_weights_0.72.hdf5', custom_objects={'KerasLayer': hub.KerasLayer})
```

## Load Data

```python
test_datagen = ImageDataGenerator()
validation_generator = test_datagen.flow_from_directory(
    'images/test',
    target_size=(img_height, img_width),
    batch_size=batch_size,
    seed = 11,
    class_mode='categorical')
 # beinhaltet array von images und array von one-hot-codierten labels
label_map = (validation_generator.class_indices) # this may be helpful for accuracy per class
```

## Calculate Accuracy

```python
correct_classifications = 0
i = 0
for images, true_labels in validation_generator:
      print(f'start prediction') # - hier kommt er schon nicht mehr an
      pred_labels = image_model.predict(images)
      batch_correct = 0
      for actual_label, label in zip(true_labels, pred_labels):
            if actual_label.argmax(0) == label.argmax(0): # d.Classes[alphabetically_to_classes[label.argmax(0)]]: # d.Classes[label.argmax(0)]:
                  correct_classifications += 1
                  batch_correct += 1
      batch_accuracy = batch_correct / batch_size
      print(f'{batch_accuracy=}')
      i += 1
      print(f'samples looked at:{i*batch_size}')
      

accuracy = float(correct_classifications) / num_test_samples
print(f'{accuracy=}')

#TODO: Why does it stop after 5 iterations!?!?!?!? - seems to be some issue with the memory! - somehow it does not stop at the end of the generator
```
